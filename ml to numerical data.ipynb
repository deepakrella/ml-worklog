# Apply lograithemtic for the skewed columns

# split the values for train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,target, test_size = 0.2, random_state=23)
```

```python
# Check the class distribution in train data
y_train.value_counts(normalize=True)
```

```python
# Check the class distribution in train data
y_test.value_counts(normalize=True)
```

# Apply GasussianNB

```python
# import the classfier 
from sklearn.naive_bayes import GaussianNB
# create an instance of the classfier 
nb= GaussianNB(random_state=23)
# Train the model with training data
nb.fit(x_train,y_train)
```

```python
# Predict the class for test data
from sklearn.metrics import f1_score
y_pred=nb.predict(x_test)
f1_score(y_pred,y_test)
```

# Creating & Training KNN Model

```python
# import the classfier 
from sklearn.neighbors import KNeighborsClassifier
# create an instance of the classfier 
knn= KNeighborsClassifier()
# Train the model with training data
knn.fit(x_train,y_train)
```

```python
# Predict the class for test data
y_pred=knn.predict(x_test)
f1_score(y_pred,y_test)
```

# Apply SGDC

```python
# import the classfier 
from sklearn.linear_model import SGDClassifier
# create an instance of the classfier 
sgdc= SGDClassifier(random_state=23)
# Train the model with training data
sgdc.fit(x_train,y_train)
```

```python
# Predict the class for test data
y_pred=sgdc.predict(x_test)
f1_score(y_pred,y_test)
```
